```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math, os, random
import scipy.stats as stat
from mpl_toolkits.mplot3d import Axes3D

sns.set_style('darkgrid')
```

```python
iris = pd.read_csv('../data/iris.csv')
```

```python
# scatter plot
plt.figure(figsize=(10,10))
plt.scatter(iris_feat['sepal.length'], iris_feat['petal.length'], alpha=1, c='red')
plt.title('Scatter Plot', fontsize=20, fontweight='bold')
plt.xlabel('Sepal Length', fontsize=12, fontweight='bold')
plt.ylabel('Petal Length', fontsize=12, fontweight='bold')
plt.show()
```

```python
# line plot
columns = iris_feat.columns
x_data = range(0, iris.shape[0])
fig, ax = plt.subplots(figsize=(12,10))
for column in columns:
    ax.plot(x_data, iris[column], label=column)
ax.set_title('Iris Dataset')
ax.legend(loc='best')
```

```python
# line graph with seaborn with all the columns
plt.figure(figsize=(12,10))
sns.lineplot(data=iris, x='sepal.length', y='sepal.width', hue='variety', markers=True, dashes=False)
```

```python
# making A histogram
fig, ax = plt.subplots(figsize=(12,10))
ax.hist(iris_feat['sepal.length'], bins=10)
ax.set_title('sepal length')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```
